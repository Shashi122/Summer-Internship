```python
import numpy as np
import pandas as pd
```

```python
#Importing dataset
df=pd.read_excel('../input/bom-data/BOM data_Final.xlsx')
df
```

```python
#Droping irrelevant columns
df=df.drop(["Level","Item Identifier","Proc type","Conversion Cost\nPer item","Total Conversion\nCost","TOTAL COST"],axis=1)
df
```

```python
#Checking of missing values
df.isnull().sum()
```

```python
#Droping rows having missing values
df = df.dropna(subset=["Material Cost\nPer item","Material Overhead\nPer item","Total Material Cost"])
df
```

```python
#Correlation matrix
corr_matrix = df.corr()
corr_matrix
```

```python
x = df.iloc[:,0:-1]#Independent variables
y = df.iloc[:,-1]#Target variable
```

```python
# Initialise the Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
x = scaler.fit_transform(x)
```

```python
# Splitting the Data set in training and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
#MULTIPLE LINEAR REGRESSION
```

```python
# Fitting multiple linear regression to the training set 
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)
#R-Square value
from sklearn.metrics import r2_score    
coefficient_of_dermination = r2_score(y_test, y_pred)
print("Score: {}".format(coefficient_of_dermination))
```

```python
#Intercept and coefficients of polynomial equation
print(regressor.intercept_)
print(regressor.coef_)
```

```python
#DECISION TREE REGRESSION
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_dt = DecisionTreeRegressor(max_depth=10,random_state=0)
regressor_dt.fit(x_train,y_train)
y_pred = regressor_dt.predict(x_test)

#R-Square value 
coefficient_of_dermination = r2_score(y_test, y_pred)
print("Score: {}".format(coefficient_of_dermination))
```
